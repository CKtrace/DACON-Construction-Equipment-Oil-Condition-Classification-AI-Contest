```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("train.csv")
df1 = pd.read_csv("test.csv")
df.head()
```

```python
df.drop(['ID'], axis=1, inplace=True)
df1.drop(['ID'], axis=1, inplace=True)
df.head()
```

```python
df = df.fillna(0)
df1 = df1.fillna(0)
df.head()
```

```python
categorical_features = ['COMPONENT_ARBITRARY', 'YEAR']
```

```python
le = LabelEncoder()

df[categorical_features[0]] = le.fit_transform(df[categorical_features[0]])
df1[categorical_features[0]] = le.fit_transform(df1[categorical_features[0]])

le1 = LabelEncoder()

df[categorical_features[1]] = le1.fit_transform(df[categorical_features[1]])
df1[categorical_features[1]] = le1.fit_transform(df1[categorical_features[1]])

df.head()
```

```python
df.to_csv("just_basic_train.csv", index=False)
df1.to_csv("just_basic_test.csv", index=False)
```
